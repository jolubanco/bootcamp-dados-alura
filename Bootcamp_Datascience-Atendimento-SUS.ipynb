## Caderno de exploração dos dados do DataSUS de 2008 até 2021 correspondente ao Valor total por Ano/mês atendimento segundo Unidade da Federação

```python
import pandas as pd
```

```python
#Os dados não utilazam o padrão UTF-8 de acentuação, assim precisamos modicar a base dados, além disso os dados possuem
#3 linhas e 12 colunas de dados 'lixo', não trazem nenhum informação, e por fim precisamos formatar como os dados serão
#apresentados utilizando o padrão americano de casas decimais e pontuação de milhares

dados = pd.read_csv("dados/base_de_dados_SUS_atendimento.csv", sep =";", encoding="ISO-8859-1",
                    skiprows = 3, skipfooter = 12, thousands=".", decimal=",")
```

```python
#Forma de plotar os dados utilizando o pandas
dados.plot(x ="Unidade da Federação", y ="2008/Ago", kind="bar", figsize=(9,6))
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
#utilizando o Matplotlib para ajustar alguns detalhes do plot
#padrão abreviar o gráfico para 'ax'

ax = dados.plot(x ="Unidade da Federação", y ="2008/Ago", kind="bar", figsize=(13,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
#formatando os valores do eixo y para ponto flutuante, com duas casas decimais e ',' como separador de milhar (padrao americano)
plt.title("Valor gasto com o SUS, em Agosto de 2008, por unidade federativa")
plt.show()
```

```python
#Gráfico utilizando mês mas recente disponível
ax = dados.plot(x="Unidade da Federação", y = "2021/Mar", kind = "bar", figsize=(13,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor gasto com o SUS, em Março de 2021, por unidade federativa")
plt.show()
```

```python
#index é uma série imutável, logo precisamos de uma lista do python que é mutável e podemos inserir dados
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0,"Unidade da Federação") #insere na posição 0 da minha lista
colunas_usaveis
```

```python
usaveis = dados[colunas_usaveis]
usaveis.head()
```

```python
usaveis = usaveis.set_index("Unidade da Federação")
usaveis.head()
```

```python
usaveis.loc["12 Acre"] #linha cuja o indice é acre
```

```python
usaveis.T.head()
```

```python
usaveis.T.plot(figsize=(12,6))
```

```python
usaveis = usaveis.drop("Total", axis = 1)
usaveis.head()
```

```python
usaveis.T.plot(figsize=(15,6))
```

```python
#plotar utilizando estados aleatórios
#setando a semente (seed) para obter os mesmos resultados, utilizando o numpy
import numpy as np
np.random.seed(687942)
usaveis.sample(5).T.plot(figsize=(12,6))
```

```python
#Adicionando uma coluna nova
usaveis = usaveis.sum(axis=1)
```
